```python
import pandas as pd

df = pd.read_csv("/content/Amazon (1).csv")
```

**Basic Cleaning**

```python
# Convert date column
df['OrderDate'] = pd.to_datetime(df['OrderDate'])

# Check nulls
print(df.isnull().sum())

# Fill missing values (if any)
df['Discount'] = df['Discount'].fillna(0)
df['Tax'] = df['Tax'].fillna(0)

# Remove duplicates
df = df.drop_duplicates()
```

**Create NEW Columns**

Time-Based Columns

```python
df['Year'] = df['OrderDate'].dt.year
df['Month'] = df['OrderDate'].dt.month
df['MonthName'] = df['OrderDate'].dt.strftime('%B')
df['Quarter'] = df['OrderDate'].dt.to_period('Q')
```

Customer Segmentation

```python
df['CustomerType'] = df['FinalAmount'].apply(lambda x: 'High Value' if x > 500 else 'Low Value')
```

Order Status Grouping

```python
df['DeliveryStatus'] = df['OrderStatus'].replace({
    'Delivered': 'Completed',
    'Shipped': 'In Transit',
    'Pending': 'Pending'
})
```

**Save Preprocessed Dataset**

```python
df.to_csv("Amazon_Preprocessed.csv", index=False)
```
